### Evaluation Metrics

To compare the predicted outcomes against actual results, a confusion matrix is used to visualize the model’s performance by showing the counts of correct predictions (True Positives and True Negatives) and incorrect predictions (False Positives and False Negatives).

In the context of churn prevention there are two incorrect scenarios that need to be prevented: 
1. False Positive :  when a customer is predicted to churn but does not actually leave 
2. False Negative : when a customer is predicted not to churn but actually leaves

In the case of customer churn prevention strategy, a false positive would mean additional cost due to the whats deemed as unnecessary marketing/retention cost. On the other hand, a false negative would result in revenue lost. Which results in needing a more rigorous customer acquisition effort to compensate the churned ones.

Since the objective of this analysis is customer churn prevention, minimizing false negatives is critical to avoid losing customers. Therefore, evaluation focuses on the F2-score, which prioritizes recall over precision. 

### Classification Metrics

Precision

Precision represents the proportion of instances predicted as churn that are correctly classified. In churn prediction, precision indicates the efficiency of retention interventions by measuring how many targeted customers would have actually churned. Low precision corresponds to a high number of false positives, meaning resources may be allocated to customers who would have remained without intervention. Although retention efforts are generally less costly than customer acquisition, a large number of false positives can still reduce cost-effectiveness and operational efficiency.



Recall (Sensitivity)

Recall, also referred to as sensitivity, measures the proportion of actual churn cases that are correctly identified by the model. Recall is particularly important in churn prediction because failure to identify at-risk customers (false negatives) may result in preventable customer loss. Since customer acquisition costs are typically higher than retention costs, false negatives may have substantial financial implications and increase the need for new customer acquisition.



F1 Score

The F1 score is the harmonic mean of precision and recall and provides a balanced assessment when both false positives and false negatives are considered important. However, in churn prediction, the misclassification costs are often asymmetric, with false negatives typically carrying higher costs than false positives. As a result, while the F1 score offers a useful overall measure, it may not fully reflect the relative importance of recall in this context.



F2 Score (Recall-Weighted Metric)

The F2 score is a variant of the F-measure that assigns greater weight to recall than precision. Specifically, recall is weighted more heavily in the harmonic mean calculation, making the F2 score more sensitive to false negatives. This characteristic makes it particularly relevant in churn prediction contexts where failing to identify customers at risk of churning is considered more costly than incorrectly targeting customers who would have remained.



Accuracy

Accuracy measures the proportion of all predictions that are correct. In churn prediction, accuracy can be misleading due to class imbalance, as the majority of customers typically do not churn. A model may achieve high accuracy by predominantly predicting the majority class while still performing poorly in identifying actual churn cases. Therefore, accuracy alone is insufficient for evaluating model performance in churn scenarios.

## Feature Selection

To further validate regarding of variables with strong influence on Churn, we will proceed our investigation with a Hypothesis Testing. Hypothesis testing is a statistical metho to determine if there's enough evidence in a sample to infer about the phenomena as a whole. In this case, by testing a claim (null hypothesis) we will analyze the data sampled, calculate their test statistic and compare the p-value to a significance level (alpha) to decide whether to reject the null hypothesis (indicating a statistically signifiant result)

### Hypothesis Testing

Based on our normality test, the numerical data mentioned do not follow a normal distribution. Since we are comparing a binary (Yes/No) on Churn to a non-normally distributed numerical or ordical dependent variable, we will proceed with the Mann-Whitney U test. 

```python
numerical_features = [
    'tenure',
    'MonthlyCharges',
    'TotalCharges',
    'ActualTotalCharges',
    'ChargeDiff'
]

results = []

for feature in numerical_features:
    churn_yes = data[data['Churn'] == 'Yes'][feature]
    churn_no  = data[data['Churn'] == 'No'][feature]

    u_stat, p_value = mannwhitneyu(churn_yes, churn_no, alternative='two-sided')

    results.append({
        'Feature': feature,
        'U Statistic': u_stat,
        'p-value': p_value,
        'Significant (α=0.05)': 'Yes' if p_value < 0.05 else 'No'
    })

mw_results_table = pd.DataFrame(results)
mw_results_table
```

Based on the hypothesis testing, the numerical features such as tenure, monthly charges, and total charge are statistically associated with churn. With their p-value shows significance, which means that for the subsequent classification models, these results support the inclusion of tenure and monthly charges.

```python
data['Churn_bin'] = data['Churn'].map({'No': 0, 'Yes': 1})

numerical_features = ['tenure', 'MonthlyCharges', 'TotalCharges', 'ActualTotalCharges','ChargeDiff']

results = []

for feature in numerical_features:
    corr, p_value = pointbiserialr(data['Churn_bin'], data[feature])

    results.append({
        'Feature': feature,
        'Correlation': corr,
        'p-value': p_value,
        'Significant (α=0.05)': 'Yes' if p_value < 0.05 else 'No'
    })

corr_table = pd.DataFrame(results)
corr_table
```

#### SHAP Probability

```python
def shap_probability_cases(pipe, X, background_size=700):

    print("Pick a data closest to 0%, 50%, and 100% probability")

    preprocessor = pipe.named_steps['preprocessor']
    model = pipe.named_steps['model']

    X_trans = preprocessor.transform(X)
    X_trans = pd.DataFrame(X_trans, columns=preprocessor.get_feature_names_out())

    probs = model.predict_proba(X_trans)[:, 1]

    idx_0   = np.argmin(np.abs(probs - 0.0))
    idx_50  = np.argmin(np.abs(probs - 0.5))
    idx_100 = np.argmin(np.abs(probs - 1.0))

    chosen = {
        "~0% Probability": idx_0,
        "~50% Probability": idx_50,
        "~100% Probability": idx_100
    }

    background = shap.sample(X_trans, background_size)

    explainer = shap.Explainer(model.predict_proba, background)

    shap.initjs()

    for label, j in chosen.items():
        print("\n============================")
        print(f"Case: {label}")
        print(f"Data: Data Number {j+1}")
        print(f"Churn Probability: {probs[j]*100:.2f}%")
        print("============================")

        shap_values = explainer(X_trans.iloc[[j]])

        display(
            shap.plots.force(
                shap_values[0][:,1],
                matplotlib=False
            )
        )
```

```python
shap_probability_cases(pipe_tuned, X_test)
```

- The (~12%) churn probability is driven by strong negative feature contributions that push the prediction well below the base value. Long-term contract types and the presence of Online Security and Tech Support act as dominant protective factors, resulting in a consistently low churn risk.
- The (~55%) churn probability reflects a balance between opposing feature contributions around the base value. Month-to-month contract status increases churn risk, while Online Security and Tech Support partially offset this effect, leading to a prediction close to the decision threshold.
- The (~85%) churn probability is driven by strong positive feature contributions that push the prediction well above the base value. The combination of a month-to-month contract, lack of Online Security and Tech Support, and Fiber Optic internet service consistently amplifies churn risk.


Churn is largely driven by contract length and service perks. Long-term contracts combined with support services reduce churn risk, whereas short-term contracts without these services lead to a higher churn risk.

# Cost and Benefit

Cost–Benefit Analysis of Churn Prediction Outcomes

To assess the economic value of the churn prediction model, a cost–benefit analysis is conducted based on the four possible outcomes of the confusion matrix. Each outcome carries different financial implications for the firm.

Assumptions

Average customer acquisition cost (CAC) in the telecommunications industry: USD 692
(First Page Sage, 2024)

Customer retention cost is estimated to be 5–25 times lower than acquisition cost
→ Approximate retention cost range: USD 28 – 138 per customer

Retention actions include targeted promotions, discounts, or customer support outreach.

Financial Impact by Prediction Outcome
1. True Positive (TP)

Customer is predicted to churn and actually churns without intervention.

In this scenario, the model correctly identifies a customer at risk. A retention effort is applied, incurring a relatively low retention cost. If successful, the company avoids losing the customer and prevents the need for costly customer acquisition.


2. True Negative (TN)

Customer is predicted not to churn and does not churn.

No retention action is taken, and the customer remains active. This outcome incurs no additional cost and reflects efficient resource allocation.


3. False Positive (FP)

Customer is predicted to churn but does not actually churn.

A retention effort is applied unnecessarily, resulting in additional marketing or incentive costs. However, this cost is limited to the relatively low retention expense and does not result in revenue loss.


4. False Negative (FN)

Customer is predicted not to churn but actually churns.

This is the most costly error. No retention action is taken, resulting in customer loss. To maintain customer base size, the firm must acquire a new customer, incurring the full acquisition cost.
A cost–benefit analysis was conducted to evaluate the economic value of the churn prediction model using the outcomes of the confusion matrix. The model classifies customers into four categories: true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN), each of which carries distinct financial implications. Industry benchmarks indicate that the average customer acquisition cost (CAC) in telecommunications is approximately USD 692, while customer retention initiatives—such as targeted promotions, discounts, or proactive customer service—typically cost between USD 28 and USD 138 per customer. Because retention is significantly less expensive than acquisition, correctly identifying customers at risk of churning has strong financial value.

Analysis: 
True positives represent customers who were predicted to churn and would indeed have churned without intervention. In practice, the model triggers a retention effort for these customers. Although this incurs a modest retention cost, it prevents the much higher expense of acquiring a replacement customer. Therefore, each true positive generates a substantial net financial benefit equal to the avoided acquisition cost minus the retention expense. This makes true positives the most valuable outcome of the model.

True negatives occur when customers are predicted not to churn and remain with the company. No retention action is taken in this case, meaning no additional cost is incurred and no extra financial benefit is generated beyond maintaining the status quo. This outcome reflects efficient allocation of retention resources, as the company avoids unnecessary spending on customers who are already likely to stay.

False positives arise when customers are predicted to churn but would have remained loyal even without intervention. These cases lead to unnecessary retention spending, resulting in a small financial loss equal to the retention cost. However, this loss is limited and manageable because retention efforts are relatively inexpensive compared to customer acquisition. From a strategic perspective, false positives represent a tolerable trade-off if they help reduce more costly errors elsewhere.

False negatives are the most financially damaging outcome. These customers are predicted not to churn but ultimately leave the company. Because no retention action is taken, the firm loses the customer and must spend the full acquisition cost to replace them. This represents a significant financial loss per case and highlights the importance of minimizing false negatives when designing and tuning a churn prediction model.

# Kesimpulan dan Saran

Based on the SHAP analysis, the most influential driver of churn is the month-to-month contract type. Customers on flexible, non-binding contracts consistently push the model toward higher churn probability, while one-year and two-year contracts reduce churn risk. This indicates that commitment length plays a critical role in retention, and customers who are not tied into longer agreements are far more likely to leave.

Another major contributor to churn is the absence of protection and support services, particularly Online Security and Technical Support. Customers who do not subscribe to these services show a strong positive impact on churn, especially those using fiber optic internet. Fiber customers typically pay more and therefore have higher service expectations; when issues arise and support is lacking, dissatisfaction escalates quickly and results in churn.

Customer tenure also affects churn, though more moderately. Customers with shorter tenure tend to push churn predictions upward, while longer-tenured customers are more stable and less likely to leave. This suggests that the early stages of the customer lifecycle represent a vulnerable period where expectations are still forming and dissatisfaction can more easily lead to cancellation.

Payment behavior provides additional behavioral signals. Customers using manual payment methods such as electronic checks or mailed checks are slightly more likely to churn, while those using automatic payments via credit card or bank transfer are more likely to stay. Automatic payment methods likely reduce friction, missed payments, and cognitive effort, all of which contribute to higher retention.

In contrast, streaming services such as Streaming TV and Streaming Movies show minimal impact on churn. Whether customers subscribe or not does not significantly shift the model’s prediction, indicating that these features are not decisive factors in customers’ decisions to stay or leave. They function more as complementary offerings rather than retention drivers.

From a business perspective, the findings suggest that churn is primarily driven by customers who combine high flexibility, higher service cost, and low perceived support. To reduce churn, the company should prioritize converting month-to-month customers into longer contracts through incentives rather than penalties. Bundling technical support and online security—especially for fiber customers—should be positioned as essential value rather than optional add-ons. Early-stage customers should receive proactive engagement during their first few months, and automatic payment methods should be encouraged through soft incentives. Investments in streaming-related perks, however, are unlikely to deliver meaningful churn reduction compared to improvements in contract structure and service support.
